```python
for x in range(1960,2016):
    
    is_1960 = df1_subset['year']==x
    df_1960 = df1_subset[is_1960 ]
    sns.scatterplot(x='fertility_rate', y='life expectancy', hue='country', size= 'population',
            data=df_1960, alpha=0.6)
    z = str(x)
    plt.axis()
    plt.savefig('lifeexp_' + z + '.png' )
plt.close() 
```

```python
import imageio

images = []

for i in range(1960, 2016):
    filename = 'lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=20)
```
